```python
import re
import matplotlib.pyplot as plt


def get_valid_acc(log_file):
    valid_acc_list = []
    log = open(log_file,'r')
    for line in log.readlines():
        match = re.findall('valid_acc \d+.\d+', line)
        if match:
            valid_acc = match[0].split(' ')[1]
            valid_acc_list.append(float(valid_acc))
        
    return valid_acc_list
```

```python
exp_train_acc = get_valid_acc('/raid/zy_lab/yzx/darts/cnn/eval-train-20200909-140711/log.txt')
adv_train_acc = get_valid_acc("/raid/zy_lab/yzx/darts/cnn/eval-train-20200909-142735/log.txt")

plt.plot(exp_train_acc, label='exp')
plt.plot(adv_train_acc, label ='adv')
plt.legend()
```

```python
exp_search_acc = get_valid_acc('/raid/zy_lab/yzx/darts/cnn/search-EXP-20200907-165306/log.txt')
adv_search_acc = get_valid_acc("/raid/zy_lab/yzx/darts/cnn/search-ADV-20200907-172904/log.txt")

plt.plot(exp_search_acc, label='exp')
plt.plot(adv_search_acc, label ='adv')
plt.legend()
```
